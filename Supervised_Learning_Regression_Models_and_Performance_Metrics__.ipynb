## Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

##Purpose of Simple Linear Regression

It helps identify whether and how strongly X affects Y.

- **To predict values**

- **To measure impact**

The slope (β₁) tells how much Y changes when X increases by 1 unit.

- **To explain trends**
## New Section

# Question 2: What are the key assumptions of Simple Linear Regression?

 -  Key Assumptions of Simple Linear Regression

Simple Linear Regression is based on the following key assumptions:

1. Linearity
There is a linear relationship between the independent variable (X) and the dependent variable (Y).

2. Independence of Errors
The residuals (errors) are independent of each other.

3. Homoscedasticity
The variance of errors is constant for all values of the independent variable.

4. Normality of Errors
The residuals are normally distributed with a mean of zero.

5. No Multicollinearity
Since Simple Linear Regression has only one independent variable, multicollinearity does not occur.

Conclusion:
If these assumptions are satisfied, Simple Linear Regression gives reliable and accurate results.



## Question 3: Write the mathematical equation for a simple linear regression model and explain each term.
: Mathematical Equation of Simple Linear Regression

The mathematical equation of a Simple Linear Regression model is:

Y = β0 + β1X + ε

Explanation of each term:

Y  = Dependent variable (the value we want to predict)
X  = Independent variable (the input variable)
β0 = Intercept (value of Y when X = 0)
β1 = Slope or coefficient (change in Y for one-unit change in X)
ε  = Error term (difference between actual and predicted values)

Conclusion:
This equation represents a straight-line relationship between the independent and dependent variables.

## Question 5: What is the method of least squares in linear regression?

: Method of Least Squares in Linear Regression

The method of least squares is a technique used in linear regression to find the best-fit line for a given set of data points.

Explanation:
This method estimates the regression coefficients by minimizing the sum of the squared differences between the actual values and the predicted values.

Purpose of Least Squares:

• It reduces prediction errors  
• It finds the most accurate line representing the data  
• It ensures the best possible fit for the regression model  

Conclusion:
The least squares method selects regression coefficients that minimize the total squared error, resulting in the optimal regression line.


# Question 6: What is Logistic Regression? How does it differ from Linear Regression?

: Logistic Regression and its Difference from Linear Regression

Logistic Regression is a statistical method used for classification problems where the dependent variable is categorical, usually binary (0 or 1).

Purpose of Logistic Regression:
It estimates the probability that an observation belongs to a particular class using the logistic (sigmoid) function.

Differences between Logistic Regression and Linear Regression:

1. Type of Output:
Linear Regression predicts continuous values.
Logistic Regression predicts probabilities and class labels.

2. Dependent Variable:
Linear Regression uses a continuous dependent variable.
Logistic Regression uses a categorical (binary) dependent variable.

3. Model Equation:
Linear Regression uses a straight-line equation.
Logistic Regression uses a sigmoid (S-shaped) curve.

4. Range of Output:
Linear Regression output ranges from (-∞, +∞).
Logistic Regression output ranges between 0 and 1.

Conclusion:
Linear Regression is used for prediction of continuous values, while Logistic Regression is used for classification problems.





# Question 7: Name and briefly describe three common evaluation metrics for regression models?

: Common Evaluation Metrics for Regression Models

Three commonly used evaluation metrics for regression models are:

1. Mean Absolute Error (MAE):
MAE measures the average of the absolute differences between actual and predicted values.
Lower MAE indicates better model performance.

2. Mean Squared Error (MSE):
MSE calculates the average of the squared differences between actual and predicted values.
It penalizes larger errors more heavily than MAE.

3. Root Mean Squared Error (RMSE):
RMSE is the square root of MSE and represents error in the same units as the dependent variable.
Lower RMSE indicates a better fit of the model.

Conclusion:
These metrics help evaluate how accurately a regression model predicts continuous values.



\Question 8: What is the purpose of the R-squared metric in regression analysis?

: Purpose of R-squared in Regression Analysis

R-squared (R²) is a statistical metric used to measure how well the independent variable(s) explain the variation in the dependent variable.

Purpose of R-squared:

1. It shows the proportion of variance in the dependent variable that is explained by the regression model.

2. It helps evaluate the goodness of fit of the model.

3. A higher R-squared value indicates that the model explains more variability in the data.

4. It is used to compare different regression models based on how well they fit the data.

Conclusion:
R-squared tells how effective the regression model is in explaining the dependent variable.


# Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
(Include your Python code and output in the code box below.)



```python
# : Simple Linear Regression using scikit-learn

# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data (Independent variable X and Dependent variable Y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

## *Question* 10: How do you interpret the coefficients in a simple linear regression model?

: Interpretation of Coefficients in Simple Linear Regression

In a simple linear regression model, the coefficients explain the relationship between the independent variable (X) and the dependent variable (Y).

Interpretation of coefficients:

1. Intercept (β0):
The intercept represents the value of the dependent variable (Y) when the independent variable (X) is equal to zero.

2. Slope (β1):
The slope indicates how much the dependent variable (Y) changes for a one-unit increase in the independent variable (X).

Conclusion:
The coefficients help understand the direction and strength of the relationship between X and Y in the regression model.
